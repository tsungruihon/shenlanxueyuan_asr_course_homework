Q：使用 SRILM 获得经过 interpolation 式的 Kneser Ney 平滑的 3 gram 以上的语言模型。总结你观察到的一些现象。

```python
# train_set: minitrain.txt, minitrain2.txt
# test_set: test1.txt, test2.txt
# 1. 从语料库中生成n-gram计数文件
# -text 指向输入文件
# -order 指向生成几元的n-gram，即n
# -write 指向输出文件
!ngram-count -text thchs30_train.txt -order 3 -write thchs30_train.txt.count

# 2. 从上一步生成的计数文件中训练语言模型
# -read 指向输入文件，为上一步的输出文件
# -order 指向生成几元的n-gram，即n
# -lm 指向训练好的语言模型输出文件
# 最后两个参数为所采用的平滑方法，-interpolate为插值平滑，-kndiscount为 modified　Kneser-Ney 打折法，这两个是联合使用的
# 当我们执行这步的时候我们会发现会出现报错
# ”““one of modified KneserNey discounts is negative
# error in discount estimator for order 1”“”
# 表示进行平滑处理的时候由于数据集太小出现了discount的值为负数导致程序退出
# 吕航老师解答
# kndiscount是modifying的KN算法，由goodman他们提出。
# -ukndiscount是原始的，他们会统计出现过N词的n-gram计数，就是good-turning里面的计数
# 然后算打折值，通常这种情况会换成written bell
# !ngram-count -read thchs30_train.txt.count -order 3 -lm LM_train -interpolate -kndiscount -debug 1
!ngram-count -read thchs30_train.txt.count -order 3 -lm LM_train

# 3. 利用上一步生成的语言模型计算测试集的困惑度
!ngram -ppl thchs30_test.txt -order 3 -lm LM_train > result_test

# 4. 为每条句子单独打分
!ngram -ppl thchs30_test.txt -order 3 -lm LM_train -debug 1 > result_test_sent
```
